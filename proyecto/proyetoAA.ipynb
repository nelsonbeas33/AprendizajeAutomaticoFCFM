Modelo supervisado

```python
data = pd.read_csv("/AA_FCFM/MyDrive/AA_FCFM/Cancer_Data.csv")
data['diagnosis'] = data['diagnosis'].map({'M': 1, 'B': 0})
data = data.drop(columns=["Unnamed: 32"])
print(data)
```

```python
data_unlabel = data.drop(columns=["diagnosis"])
lable = data["diagnosis"]
```

```python
k_best = SelectKBest(score_func=f_regression, k = 16)
fit = k_best.fit(data_unlabel, lable)
data_unlabel.columns[fit.get_support(indices=True)].tolist()
```

```python
data[data_unlabel.columns[fit.get_support(indices=True)].tolist()]
```

```python
MatrixData = data[data_unlabel.columns[fit.get_support(indices=True)].tolist()].to_numpy().reshape(len(data),4, 4)

trainMatrixData = MatrixData[:math.floor(len(MatrixData)*0.8)]
testMatrixData = MatrixData[math.floor(len(MatrixData)*0.8):]

trainLable = data["diagnosis"][:math.floor(len(data)*0.8)]
testLable = data["diagnosis"][math.floor(len(data)*0.8):]
```

```python
modelo = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(4, (2, 2), activation='relu', input_shape=(4, 4, 1)),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(8, activation='relu'),
    tf.keras.layers.Dense(1, activation='softmax')
])

modelo.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
historial = modelo.fit(x = trainMatrixData, y = trainLable, epochs=8)
```

```python
results = modelo.evaluate(testMatrixData, testLable, batch_size=8)
print("test loss, test acc:", results)
```

```python
nuevoDatoSupervisado = pd.DataFrame({
    'radius_mean': [0.0],
    'perimeter_mean': [0.0],
    'area_mean': [0.0],
    'compactness_mean': [0.0],
    'concavity_mean': [0.0],
    'concave points_mean': [0.0],
    'radius_se': [0.0],
    'perimeter_se': [0.0],
    'area_se': [0.0],
    'radius_worst': [0.0],
    'texture_worst': [0.0],
    'perimeter_worst': [0.0],
    'area_worst': [0.0],
    'compactness_worst': [0.0],
    'concavity_worst': [0.0],
    'concave points_worst': [0.0]
})

nuevoDatoSupervisado = nuevoDatoSupervisado.to_numpy().reshape(len(nuevoDatoSupervisado),4, 4)
```

```python
predictions = modelo.predict(nuevoDatoSupervisado)
predictions = [0 if pred > 0.5 else 1 for pred in predictions]
print(predictions)
```

```python
data = pd.read_csv("/AA_FCFM/MyDrive/AA_FCFM/Cancer_Data.csv")
data = data.drop(columns=['Unnamed: 32'])
data['diagnosis'] = data['diagnosis'].replace({'M': 1, 'B': 0})
print(data)
```

```python
subset_columns = ['diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean']
data = data[subset_columns]
print(data)
```

```python
EtiquetaArbol = data["diagnosis"]
datosArbol = data.drop(columns=["diagnosis"])
```

```python
trainDatosArbol = datosArbol[:math.floor(len(datosArbol)*0.8)]
testDatosArbol = datosArbol[math.floor(len(datosArbol)*0.8):]

trainEtiquetaArbol = EtiquetaArbol[:math.floor(len(EtiquetaArbol)*0.8)]
testEtiquetaArbol = EtiquetaArbol[math.floor(len(EtiquetaArbol)*0.8):]
```

```python
from sklearn.tree import DecisionTreeClassifier

arbol = DecisionTreeClassifier()
arbol.fit(trainDatosArbol, trainEtiquetaArbol)
```

```python
arbol.score(testDatosArbol, testEtiquetaArbol)
```
